# pyplot

```python
import numpy as np
import matplotlib.pyplot as plt    # 导入集成的pyplot
```

    plt.show()   # 生成图像，默认弹窗
    所以要加上 %matplotlib inline  这样图像会展示在jupyter的页面内部

```python
%matplotlib inline
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel("y number")    # y 坐标的名称

plt.show()
```

```python
plt.plot([1,2,3], [5,7,4])
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16])      # y=x^2   
```

```python
# 常见写法
x = np.linspace(1, 20, 1000)
y = x ** 2
plt.plot(x, y)
```

    画多条曲线

```python
a = np.arange(10)
plt.plot(a, a*1.5, a, a*2.5, a, a*3.5, a, a*10)   # 多个x,y 连着写
plt.show()
```

    图例、标题、和标签

```python
x = [1, 2, 3]
y = [5, 7, 4]
x2 = [1, 2, 3]
y2 = [10, 14, 12]
```

```python
plt.plot(x,y, label='F1 line')
plt.plot(x2,y2, label='F2 line')
plt.legend()   # 使用plt.legend()生成默认图例
plt.show()
```
