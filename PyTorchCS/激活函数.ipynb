## sigmoid
- 激活函数作用于每一个神经元的输出。不会改变input向量的形状。

```python
x = torch.rand(1, 10, 16)
```

```python
x
```

```python
y = F.sigmoid(x)
y
```

```python
y.size()
```

```python
m = nn.Sigmoid()
inputs = torch.randn(2)
inputs
```

```python
outputs = m(inputs)
```

```python
outputs
```

```python
def sigmoid(x):
    return 1./ (1. + torch.exp(-x))
```

```python
sigmoid(inputs)
```
